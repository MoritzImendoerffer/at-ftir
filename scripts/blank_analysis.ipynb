```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

# Summary

Quick look at Blank data

```python
# Load in all blank dataframes
b1 = pd.read_pickle('171113_Blank Dynamic_raw.p')
b1.reset_index(inplace=True)
b2 = pd.read_pickle('171114_Blank Dynamic_raw.p')
b2.reset_index(inplace=True)
b3 = pd.read_pickle('171115_Blank Dynamic_raw.p')
b3.reset_index(inplace=True)

# make one single dataframe
df = b1.append(b2)
df = df.append(b3)
df.head()
```

## Plot of the data

```python
gr = df.groupby(by='date')

fig, ax = plt.subplots(nrows=len(gr), figsize=(8, 8*len(gr)))
fig.subplots_adjust(hspace=0.5)
for index, item in enumerate(gr):
    name, val = item
    
    x = val.time
    y = val.iloc[:, 3:].T.plot(cmap=plt.cm.viridis, ax=ax[index])
    ax[index].legend(('First: dark blue', ))
    ax[index].set_title(name)
```

```python
dates = df.date.unique()
mask = df.date == dates[0]

d1 = df.loc[mask,:].drop(['date', 'name'], axis=1)
d1.set_index('time', inplace=True)
d1.head()
```
