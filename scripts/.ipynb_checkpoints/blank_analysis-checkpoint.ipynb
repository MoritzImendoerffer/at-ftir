```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy import signal
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from matplotlib.colors import Normalize
%matplotlib inline
```

# Summary

Quick look at Blank data

```python
data = pd.read_pickle('../export/one_to_rule_them_all.p')
data.name = data.name.str.lower()
data.reset_index(inplace=True)
data.head()
```

## Plot of the spectra of all blanks

```python
mask = data.name.str.contains('blank')
gr = data[mask].groupby(by='date')

fig, ax = plt.subplots(nrows=len(gr), figsize=(8, 8*len(gr)))
fig.subplots_adjust(hspace=0.5)


for index, item in enumerate(gr):
    name, val = item
    
    step = 50
    x = val.time[::step].values
    y = val.iloc[::step, 3:].values
    
    color = plt.cm.viridis(np.linspace(0,1, len(y)))
    for i, yl in enumerate(y):
        ax[index].plot(yl, label=x[i], color=color[i])
    
    
    
    
    ax[index].legend()
    ax[index].set_title(name)
    
plt.savefig('../plots/all_full_range_selected_time.png', dpit=190)
```

# PCA of preprocessed signal

```python
data.head()
```

```python
spectra = data.iloc[:,3:]
mask_wave = (spectra.columns > 1600) & (spectra.columns < 1800)
spectra = spectra.loc[:, mask_wave]
info = data.iloc[:,0:3]

spec_norm = StandardScaler().fit_transform(spectra)

spectra_to_analyze = spec_norm

pca = PCA(n_components=3)
pca.fit(spectra_to_analyze)
explained_variance = pca.explained_variance_ratio_
components = pca.transform(spectra_to_analyze)

df_pca = pd.DataFrame(components, columns=['PCA0', 'PCA1', 'PCA2'])
df_pca = pd.concat((info, df_pca), axis=1)
df_pca.time = [float(item) for item in df_pca.time]



gr = df_pca.groupby('date')

color = plt.cm.tab10

fig, ax = plt.subplots(ncols=1, figsize=(10,10))
for index, item in enumerate(gr):
    name, val = item
    
    x = val.PCA0
    y = val.PCA1
    c = val.time
    
    if any(val.name.str.contains('blank')):
        marker = 'x'
        color = plt.cm.tab20c_r
    else:
        marker = '.'
    ax.scatter(x,y, c=color(index), label=name, marker=marker)
    ax.plot(x.values[-1], y.values[-1], marker='o', color=color(index), markersize=20)

ax.legend(loc='upper left', bbox_to_anchor=(1,1))
```
